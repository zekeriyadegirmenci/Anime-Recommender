## Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(animes,on='anime_id')
ratings_with_name.head()
```

```python
num_rating_df = ratings_with_name.groupby('English name').count()['rating'].reset_index()
num_rating_df.rename(columns={'rating':'total_num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('English name').agg({'rating': 'mean'}).reset_index()
avg_rating_df.rename(columns={'rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='English name')
popular_df
```

```python
popular_df = popular_df[popular_df['total_num_ratings']>=1000].sort_values('avg_rating',ascending=False).head(50)
popular_df
```

```python
popular_df = popular_df.merge(animes,on='English name').drop_duplicates('English name')[['English name','Name','Score','Image URL']]
```

```python
popular_df
```

## Collaborative Filtering Based Recommender System

```python
#kullanıcıların en az 400 animeye oy vermiş olanlar
x = ratings_with_name.groupby('user_id').count()['rating'] > 400
most_review_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['user_id'].isin(most_review_users)]
filtered_rating
```

```python
#100den fazla degerlendirme yapılan animeler    
y = filtered_rating.groupby('English name').count()['rating']>=100
famous_animes = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['English name'].isin(famous_animes)]
```

```python
pivot_table = final_ratings.pivot_table(index='English name',columns='user_id',values='rating')
```

```python
pivot_table.fillna(0,inplace=True)
```

```python
pivot_table
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pivot_table)
```

```python
similarity_scores.shape
```

```python
def recommend(anime_name):
    # index fetch
    index = np.where(pivot_table.index==anime_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:9]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = animes[animes['English name'] == pivot_table.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('English name')['English name'].values))
        item.extend(list(temp_df.drop_duplicates('English name')['Other name'].values))
        item.extend(list(temp_df.drop_duplicates('English name')['Image URL'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('Attack on Titan')
```
